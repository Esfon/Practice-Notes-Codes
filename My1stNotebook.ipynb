# Variables

```python
# Enter Variables

x = 'Hello'
y = 4
z = 3.14

print(x)
print(y)
print(z)

# Check var type

print(type(x))
print(type(y))
print(type(z))
```

```python
# Naming variable

camelCaseVar = 'Fruit Juice'
PascalCaseVar = 'Fruit Juice'
snake_case_var = 'Fruit Juice'

print(camelCaseVar)
print(PascalCaseVar)
print(snake_case_var)
```

```python
# Inside/Outside variable(1)

x = 'I did'

def func():
    print('Quote says, ' + x)
func()
```

```python
# Inside/Outside variable(2)

x = 'I can'

def func():
    x = 'I cannot'
    print('Quote says, ' + x)
func()

print('Quote says, ' + x)
```

```python
# Inside/Outside variable(3) Global func

x = 'I won'

def func():
    global x
    x = 'I lost'
func()

print('Quote says, ' + x)
```

# Data types

```python
a = 'text' # str
b = 6      # num
c = 4.7    # num
d = 3j     # num
e = ['Mango', 'Apple', 'Banana']   #seq
f = ('Mango', 'Apple', 'Orange')   #seq
g = range(5)  #seq
h = {'Fruit': 'Orange', 'Number': 5, 'Social Media': ['Facebook', 'Twitter'], 'Course': 'Mathematics'}   # map
i = {2,3,5,7,11}   # set


print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
print(type(f))
print(type(g))
print(type(h))
print(type(i))
```

# Math/Operators

```python
x = 2
y = 11

# Add

print (x + y)

# Subtract

print (y - x)

# Multiply

print (x * y)

# Divide

print (y / x)

# Modulus/Remainder

print (y % x)

# Exponential

print (y ** x)

# Floor div

print (y // x)
```

# Pandas

```python
# Code intro with example

import pandas as pd

data = {'Country': ['PHI', 'IND', 'FRA', 'JAP', 'ITA'],
        'Number': [1, 2, 3, 4, 5]}

df = pd.DataFrame(data)

print(df)
```

```python
# Pandas version

print(pd.__version__)
```

```python
# Series

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
# Labels in series (index)

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
# Series keys/values

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```
